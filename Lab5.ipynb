```python
#3
import numpy as np
c = [3, 7, 15, 1, 292, 1, 1, 1, 2, 1, 3, 1, 14, 2] 
M = np.asmatrix([[1,2],[3,4]])
for i in range(len(c)):
    ci = np.asmatrix([[1,1],[1,0]])
    ci[0, 0] = c[i]
    if (i==0):
        M = ci
    else:
        M = M.dot(ci)
print(M)
print(M[0,0]/M[1,0]) 
```

```python
#4
import numpy as np
X = np.array([
    [1, 0],
    [1, 1],
    [1, 2],
    [1, 3],
    [1, 4]
])
Y = np.array([
    [1],
    [2],
    [3],
    [4],
    [5]
])
XT = X.T
A1 = np.linalg.inv(XT.dot(X))
A2 = XT.dot(Y)
A = A1.dot(A2)
print("Hệ số hồi quy (a0, a1):")
print(A)
f = X.dot(A)
sai_so_binh_phuong = np.sum((f - Y)**2)
print("\nTổng bình phương sai số:", sai_so_binh_phuong)
print("\nPhương trình gần đúng: y = {:.3f} + {:.3f}x".format(A[0,0], A[1,0]))
```

#Bai Tap Chuong

```python
#BT1
import numpy as np
import matplotlib.pyplot as plt
def bpcuctieu(a):
    b = []   
    o1 = []  
    for i in range(len(a)):
        o1.append(1.0)
        b.append(i * 1.0)
    x = np.array([o1, b])
    x = x.T 
    y = np.array(a).reshape(-1, 1) 
    u = np.dot(x.T, x)
    u_1 = np.linalg.inv(u)
    v = np.dot(x.T, y)
    A = np.dot(u_1, v)
    return A
dulieu = [10.0, 8, 7, 5, 2]  
A = bpcuctieu(dulieu)
print("Phương trình gần đúng: y = {:.2f} + {:.2f}x".format(A[0, 0], A[1, 0]))
thoi_gian = np.arange(len(dulieu))       
du_doan = A[0, 0] + A[1, 0] * thoi_gian   
plt.plot(thoi_gian, dulieu, 'ro-', label='Dữ liệu thực tế')
plt.plot(thoi_gian, du_doan, 'b--', label='Hồi quy tuyến tính')
plt.title("Giảm lượng thuốc theo thời gian")
plt.xlabel("Thời gian (giờ)")
plt.ylabel("Lượng thuốc (đơn vị)")
plt.legend()
plt.grid(True)
plt.show()

```
